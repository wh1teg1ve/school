```python
import pandas as pd
import numpy as np
import re
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# 读取数据
df = pd.read_csv('E_commerce.csv')

print(f"数据集形状: {df.shape}")
print(f"数据集列名: {df.columns.tolist()}")
print("\n前几行数据:")
print(df.head())
```

```python
# 1. 检查数据基本情况
print("=== 数据基本情况 ===")
print(f"总行数: {df.shape[0]}")
print(f"总列数: {df.shape[1]}")
print(f"唯一客户数: {df['Customer ID'].nunique()}")
print(f"唯一订单数: {df['Order ID'].nunique()}")
print(f"数据时间范围: {df['Order Date'].min()} 到 {df['Order Date'].max()}")
```

```python
# 3. 检查和处理缺失值
print("\n=== 缺失值检查 ===")
missing_values = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100
missing_df = pd.DataFrame({'缺失数量': missing_values, '缺失百分比': missing_percent})
print(missing_df[missing_df['缺失数量'] > 0])

# 检查特定列中的异常值（如Quantity列中的非数字值）
print("\n=== 检查Quantity列的异常值 ===")
# 查看Quantity列中的非数字值
quantity_issues = df[~df['Quantity'].apply(lambda x: str(x).replace('.', '').isdigit() if pd.notnull(x) else False)]
print(f"Quantity列非数字值数量: {quantity_issues.shape[0]}")
if quantity_issues.shape[0] > 0:
    print("异常值示例:")
    print(quantity_issues[['Quantity', 'Sales', 'Order ID']].head())
```

```python
# 4. 数据格式清洗和转换
print("\n=== 数据格式清洗 ===")

# 清洗Sales列（去除$符号和逗号）
def clean_currency(value):
    if pd.isnull(value):
        return np.nan
    if isinstance(value, str):
        # 去除$符号、逗号和空格
        cleaned = re.sub(r'[$,]', '', value.strip())
        try:
            return float(cleaned)
        except:
            return np.nan
    return float(value)

# 清洗Profit列
def clean_profit(value):
    if pd.isnull(value):
        return np.nan
    if isinstance(value, str):
        # 去除$符号、逗号和空格
        cleaned = re.sub(r'[$,]', '', value.strip())
        try:
            return float(cleaned)
        except:
            return np.nan
    return float(value)

# 清洗Shipping Cost列
def clean_shipping(value):
    if pd.isnull(value):
        return np.nan
    if isinstance(value, str):
        # 去除$符号和空格
        cleaned = re.sub(r'[$,]', '', value.strip())
        try:
            return float(cleaned)
        except:
            return np.nan
    return float(value)

# 清洗Quantity列
def clean_quantity(value):
    if pd.isnull(value):
        return np.nan
    if isinstance(value, str):
        # 尝试转换为数字
        try:
            return float(value)
        except:
            return np.nan
    return float(value)

# 应用清洗函数
df['Sales'] = df['Sales'].apply(clean_currency)
df['Profit'] = df['Profit'].apply(clean_profit)
df['Shipping Cost'] = df['Shipping Cost'].apply(clean_shipping)
df['Quantity'] = df['Quantity'].apply(clean_quantity)

# 处理Discount列（已经是小数，但确保是数值型）
df['Discount'] = pd.to_numeric(df['Discount'], errors='coerce')

# 处理日期列
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')

print("数据清洗完成!")
print(f"Sales列范围: {df['Sales'].min():.2f} - {df['Sales'].max():.2f}")
print(f"Profit列范围: {df['Profit'].min():.2f} - {df['Profit'].max():.2f}")
print(f"Quantity列范围: {df['Quantity'].min():.2f} - {df['Quantity'].max():.2f}")
```

```python
# 5. 处理缺失值
print("\n=== 处理缺失值 ===")

# 检查清洗后的缺失值
missing_after_clean = df.isnull().sum()
missing_cols = missing_after_clean[missing_after_clean > 0]

if len(missing_cols) > 0:
    print("清洗后的缺失值情况:")
    for col, count in missing_cols.items():
        percent = (count / len(df)) * 100
        print(f"{col}: {count} 行 ({percent:.2f}%)")
    
    # 处理Quantity列的缺失值（用中位数填充）
    if 'Quantity' in missing_cols:
        quantity_median = df['Quantity'].median()
        df['Quantity'] = df['Quantity'].fillna(quantity_median)
        print(f"Quantity列用中位数({quantity_median})填充")
    
    # 处理Discount列的缺失值（用0填充）
    if 'Discount' in missing_cols:
        df['Discount'] = df['Discount'].fillna(0)
        print("Discount列用0填充")
    
    # 删除Order Date为空的记录（数量应该很少）
    df = df.dropna(subset=['Order Date'])
    print(f"删除Order Date为空的记录后数据形状: {df.shape}")
else:
    print("没有缺失值需要处理")
```

```python
# 6. 检查和处理异常值
print("\n=== 检查异常值 ===")

# 检查数值列的统计信息
numeric_cols = ['Sales', 'Quantity', 'Discount', 'Profit', 'Shipping Cost', 'Browsing Time (min)']
print("数值列统计信息:")
print(df[numeric_cols].describe())

# 检查负值的Profit（可能表示亏损）
negative_profit = df[df['Profit'] < 0]
print(f"\n负利润订单数: {negative_profit.shape[0]} ({negative_profit.shape[0]/len(df)*100:.2f}%)")

# 检查负值的Sales（不应该有）
negative_sales = df[df['Sales'] < 0]
print(f"负销售额订单数: {negative_sales.shape[0]}")

# 检查Quantity为0或负值的情况
invalid_quantity = df[df['Quantity'] <= 0]
print(f"无效数量(<=0)订单数: {invalid_quantity.shape[0]}")

# 处理Quantity为0或负值的情况（用1替换）
if invalid_quantity.shape[0] > 0:
    df.loc[df['Quantity'] <= 0, 'Quantity'] = 1
    print("已将Quantity<=0的值替换为1")
```

```python
# 7. 数据一致性检查
print("\n=== 数据一致性检查 ===")

# 检查年龄的合理性
invalid_age = df[(df['Age'] < 0) | (df['Age'] > 100)]
print(f"不合理年龄(小于0或大于100)记录数: {invalid_age.shape[0]}")

# 检查浏览时间的合理性
invalid_browsing = df[df['Browsing Time (min)'] < 0]
print(f"负浏览时间记录数: {invalid_browsing.shape[0]}")

# 检查折扣的合理性
invalid_discount = df[(df['Discount'] < 0) | (df['Discount'] > 1)]
print(f"不合理折扣(小于0或大于1)记录数: {invalid_discount.shape[0]}")

# 处理不合理的折扣值
if invalid_discount.shape[0] > 0:
    df.loc[df['Discount'] < 0, 'Discount'] = 0
    df.loc[df['Discount'] > 1, 'Discount'] = 1
    print("已将不合理折扣值调整到[0,1]范围内")

print("\n数据清洗完成!")
print(f"最终数据形状: {df.shape}")
print(f"数据时间范围: {df['Order Date'].min()} 到 {df['Order Date'].max()}")
```
